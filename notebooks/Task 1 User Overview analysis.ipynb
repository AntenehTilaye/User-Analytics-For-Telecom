> in order to identify the use we can use international mobile subscriber identity (IMSI)

```python
agg_data = pd.DataFrame()
```

```python

agg_data['IMSI'] = db.groupby('IMSI').groups.keys()
# agg_data['IMSI'] = db.groupby('IMSI')['IMSI']
agg_data['number_of_xdr_session'] = pd.DataFrame(db.groupby('IMSI')['IMSI']
     .agg(number_of_xdr_session = 'count')).reset_index()['number_of_xdr_session']


agg_data['session_duration'] = pd.DataFrame(db.groupby('IMSI')['Dur. (ms)']
     .agg(session_duration = 'sum')).reset_index()['session_duration']

column_names = [('Social Media UL (Bytes)', 'Social Media DL (Bytes)'),
               ('Google DL (Bytes)', 'Google UL (Bytes)'), ('Email DL (Bytes)', 'Email UL (Bytes)'), 
                ('Youtube DL (Bytes)','Youtube UL (Bytes)'), ('Netflix DL (Bytes)', 'Netflix UL (Bytes)'),
               ('Gaming DL (Bytes)', 'Gaming UL (Bytes)'), ('Other DL (Bytes)', 'Other UL (Bytes)'),
               ('Total UL (Bytes)', 'Total DL (Bytes)')]

new_names = [ 'social_total', 'google_total', 'email_total', 'youtube_total', 'netflix_total', 'gaming_total',
             'other_total', 'total_dl_and_ul_data']


for ind in range(len(column_names)):
    
    agg_data['total_dl_data'] = pd.DataFrame(db.groupby('IMSI')[column_names[ind][0]]
         .agg(total_ul_data = 'sum')).reset_index()['total_ul_data'] 
    agg_data['total_ul_data'] = pd.DataFrame(db.groupby('IMSI')[column_names[ind][1]]
                 .agg(total_dl_data = 'sum')).reset_index()['total_dl_data']

    agg_data[new_names[ind]] = agg_data[['total_dl_data','total_ul_data']].sum(axis=1)


agg_data.drop(columns=['total_dl_data', 'total_ul_data'])

agg_data.head()
```

```python
Q1 = np.percentile(agg_data['session_duration'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(agg_data['session_duration'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", agg_data.shape)
 
# Upper bound
upper = np.where(agg_data['session_duration'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(agg_data['session_duration'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
agg_data.drop(upper[0], inplace = True)
agg_data.drop(lower[0], inplace = True)

print("New Shape: ", agg_data.shape)
```

####  Variable transformations
segmenting the users into top five decile classes based on the total duration for all sessions and compute the total data (DL+UL) per decile class.

```python
# Sorting the DataFrame in descending Order of total duration
agg_data.sort_values(by =['session_duration'], ascending=False, inplace = True)
```

```python
agg_data['decile_rank']= pd.qcut(agg_data['session_duration'],
                             q = 10, labels = False)
```

```python
agg_data.head()
```

```python
# get the top five decile based on total session duration
top_five_decile = agg_data[agg_data['decile_rank'] > 4]
```

```python
top_five_decile.tail()
```

```python
# compute the total data per decile class
agg_data_segment = pd.DataFrame()
agg_data_segment['decile_rank'] = top_five_decile.groupby('decile_rank').groups.keys()

agg_data_segment['total_data_per_decile'] = pd.DataFrame(top_five_decile.groupby('decile_rank')['total_dl_and_ul_data']
     .agg(total_data = 'sum')).reset_index()['total_data']
agg_data_segment.head()
```

```python
agg_data_segment['total_data_per_decile'].plot()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

from sklearn.decomposition import PCA

lb = LabelEncoder() 
sc = StandardScaler()

col_names = ['Start', 'End', 'Last Location Name',  'Handset Manufacturer', 'Handset Type']
cat_data = db[col_names]
    

for_PCA = db.drop(columns=col_names)
for_PCA.head()
```

```python
lb.fit_transform(cat_data['Start'])
```

```python
for name in col_names:
   
    for_PCA[name] = lb.fit_transform(cat_data[name])
```

```python
for_PCA = sc.fit_transform(for_PCA)
```

```python
pca = PCA(n_components=5)
principalComponents  = pca.fit_transform(for_PCA)
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['component 1', 'component 2', 'component 3', 'component 4', 'component 5'])
principalDf.head()
```
